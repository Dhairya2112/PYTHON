```python
import math
class circle:
    def __init__(self,pie,x):
        self.pie=pie
        self.x=x
        self.peri(pie,x)
        self.area(pie,x)
    def peri(self,pie,x):
        self.pie=pie
        self.x=x
        print("perimeter",2*pie*x)
    def area(self,pie,x):
        self.pie=pie
        self.x=x
        print("perimeter",pie*(x**2))
pie=math.pi
d1=circle(pie,10)
```
