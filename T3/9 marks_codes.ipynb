read file and when "#" appears then after that line willl be commented and non commented file

```python
r=open("file1.txt","w+")
r.write("hi i am dhairya\n #tari namaskar \n i am good")
r.seek(0)
r1=r.readlines()
print(r1)
q=open("file2.txt","w+")
for i in r1:
    if i.startswith("#"):
        continue
    else:
        if "#" in i:
            ind=i.find("#")
            q.write(i[0:ind]+"\n")
        else:
            q.write(i)
q.seek(0)
q1=q.readlines()
print(q1)
r.close()
q.close()
```

find min and max bid price of players in players.txt then display it

```python
r=open("players.txt")
r1=r.readlines()
d={}
for i in r1:
    r2=i.split()
    d.update({r2[0]:int(r2[1])})
print(d)
p2=max(d.values())
for i,j in d.items():
    if j==p2:
        print(f"player-{i} with highest bid price-{j}")
p3=min(d.values())
for i,j in d.items():
    if j==p3:
        print(f"player-{i} with min bid price-{j}")
```
