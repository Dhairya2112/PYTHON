```python
# Q.1:read filembox-short and find all keyword X-DSPAM-Confidence:
# find avg of all value
x=open("mbox-short.txt")
f=x.readlines()
c=0
s=0
for i in f:
    if "X-DSPAM-Confidence:" in i:
        c=c+1
        p=i.split(" ")
        s=s+float(p[1])
print("AVG=",s/c)
```

```python
# Q.2:find all line that start from "from":
c1=0
for i in f:
    if "From " in i:
        c1=c1+1
        p=i.split(" ")
        print(p[1])
print("count:",c1)
```

```python
# Q.3 find number of times email send by email 
c1=0
c2={}
z=[]
for i in f:
    if "From " in i:
        c1=c1+1
        p=i.split(" ")
        z.append(p[1])
        if p[1] in z:
            c2.update({p[1]:z.count(p[1])})
m=max(c2.values())
for i,j in c2.items():
    print(i,":-",j)
#     
for i1,j1 in c2.items():    
    if j1==m:
        print()
        print(f"email:{i1} and number of times={j1}")
```
