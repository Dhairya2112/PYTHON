```python
class demo(Exception):
    pass
try:
    n=int(input("Enter your phone number="))
    z=str(n)
    if len(z)!=10:
        raise demo("please input 10 numbers")
    elif z[0]!="9":
        raise demo("first digit should be 9")
    else:
        print("Success")
except demo as d:
    print(d)
```
