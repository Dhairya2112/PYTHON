```python
import numpy as np
p=["T1","T2","T3","T4"]
d=[10,20,15,20]
e=[13,15,18,15]
f=[1,23,12,13]
plt.subplot(2,2,1)
plt.plot(p,d,marker="*",ms=15,mec="red",mfc="yellow",color="black",linewidth=2,linestyle="-.")
plt.subplot(2,2,2)
plt.plot(p,e,marker="*",ms=15,mec="blue",mfc="yellow",color="pink",linewidth=3,linestyle="-")
plt.subplot(2,2,3)
plt.plot(p,f,marker="*",ms=15,mec="black",mfc="yellow",color="green",linewidth=4,linestyle=":")
plt.grid(color="orange",linewidth=2)
plt.show()
```

Scatter

```python
p=["T1","T2","T3","T4"]
d=[10,20,15,20]
e=[13,15,18,15]
f=[1,23,12,13]
c1=["red","blue","purple","yellow"]
s1=[400,950,700,250]
plt.scatter(p,d,c=c1,alpha=0.4,marker="*")
plt.scatter(p,e,c=c1,alpha=0.5,marker="*")
plt.scatter(p,f,c=c1,alpha=0.7,marker="*")
plt.show()
```

```python
p=["T1","T2","T3","T4"]
d=[10,20,15,20]
plt.bar(p,d,color="blue")
plt.show()
```

```python
p=["T1","T2","T3","T4"]
d=[10,20,15,20]
plt.barh(p,d,color="blue")
plt.show()
```

```python
p=["T1","T2","T3","T4"]
d=[10,20,15,20]
plt.pie(d,labels=p)
plt.show()
```

```python
p=["T1","T2","T3","T4"]
d=[10,20,10,20]
plt.pie(d,labels=p,startangle=90)
plt.legend(title="Exam")
plt.show()
```

```python
p=["T1","T2","T3","T4"]
d=[10,20,10,20]
e=[0.2,0,0.4,0]
plt.pie(d,labels=p,startangle=90,explode=e,autopct="%1.1f") #autopct=printing percentage
plt.legend(title="Exam")
plt.show()
```

Fill Between

```python
import numpy as np
import matplotlib.pyplot as plt
x=np.array([1,2,3,4,8])
y=np.array([1,8,2,4,5])
z=np.array([1,2,5,7,3])
plt.fill_between(x,y)
plt.fill_between(x,z)
plt.show()
```

StackPlot

```python
x=np.array([1,2,3,4,8])
y=np.array([1,8,2,4,5])
z=np.array([1,2,5,7,3])
plt.stackplot(x,y,z,labels=["A","B"])
plt.legend()
plt.show()
```
