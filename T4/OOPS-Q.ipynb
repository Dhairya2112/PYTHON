```python
from abc import ABC,abstractmethod
class defence(ABC):
    @abstractmethod
    def gun(self):
        pass
    def show(self):
        print("hi")
```

```python
class Army(defence):
    def gun(self):
        print("AK47")
a=Army()
a.gun()
a.show()
```

```python
class sArmy(defence):
    def gun(self):
        print("AUG")
a=sArmy()
a.gun()
a.show()
```

```python
from abc import ABC,abstractmethod
class Employee(ABC):
    id=int(input("Enter id="))
    name=input("Enter name=")
    salary=int(input("Enter salary="))
    @abstractmethod
    def emp_id(self):
        self.id=id
        self.salary=salary
        self.name=name
    def details(self):
        print(f"id={self.id} and name={self.name}")
    def salry(self):
        print("salary=",self.salary)
class Perks(Employee):
    def emp_id(self):
        self.id=id
        self.name=name
        self.salary=salary
    def da(self):
        self.da=self.salary*0.35
        self.hra=self.salary*0.17
        self.pf=self.salary*0.12
        self.tsal=self.salary+self.da+self.hra-self.pf
        print("DA=",self.da)
        print("HRA=",self.hra)
        print("PF=",self.pf)
        print("Total Salary=",self.tsal)
p=Perks()
p.details()
p.salry()
p.da()
```
