```python
dic={1000:"M",900:"CM",500:"D",400:"CD",100:"C",90:"XC",50:"L",40:"XL",10:"X",9:"IX",5:"V",4:"IV",1:"I"}
val=[]
s=[]
for i,j in dic.items():
    val.append(i)
    s.append(j)
ans=""
i=0
n=int(input("Enter number : "))
num=n
while n>0:
    for j in range(n//val[i]):
        ans=ans+s[i]
        n=n-val[i]
    i=i+1
print(f"{num} into roman is : {ans}")
```
