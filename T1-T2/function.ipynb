```python
# map() 
# add 10 at all index and print in tuple
l1=[1,2,3,4]
z=tuple(map(lambda i:i+10,l1))
print(z) # (11, 12, 13, 14)
```

```python
# filter()
# print even numbers and print in list
l=[1,2,3,4]
z=list(filter(lambda i:i%2==0,l))
print(z) # [2, 4]
```

```python
# reduce()
# compulsury import library for this function
from functools import *
l=[1,2,3,4,5]
z=reduce(lambda x,y:x*y,l)
print(z) # 120
```

```python
# maketrans 
# to replace multiple value.

s="hello good morning"
ans=str.maketrans("ohg","@#$")
sol=s.translate(ans)
print(sol) # #ell@ $@@d m@rnin$

#ex2
w=str.maketrans("ohg","@#$","n")
v1=s.translate(w)
print(v1) # #ell@ $@@d m@ri$

# ex3
x=str.maketrans(" "," "," ")
v2=s.translate(x)
print(v2) # hellogoodmorning

# ex4
s="he@#. INDIA is =.912"
ans=s.maketrans("I","i","@#.=")
n=s.translate(ans)
print(n) # he iNDiA is 912
```
